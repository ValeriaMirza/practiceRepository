#**Lab 4 Assignment: Data Preparation with Pandas**

Importing  the datset published by the New York City Department of Parks and Recreation (CC0 1.0 Universal Public Domain Dedication) with the command below:

```python
import pandas as pd
```

```python
lab1_df=pd.read_csv('https://data.cityofnewyork.us/api/views/5rq2-4hqu/rows.csv?accessType=DOWNLOAD')
```

Using head() to view just the beginning of the file.

```python
lab1_df.head()
```

Creating a Python list containing the name of the columns of interest. Then, creating a new DataFrame called 'lab1_df_reduced' that only contains those columns.

```python
list_reduced = ['spc_latin', 'spc_common', 'status', 'health', 'tree_dbh', 'stump_diam']
lab1_df_reduced = lab1_df[list_reduced]
lab1_df_reduced
```

Scanning the new DataFrame for NaN values that may hinder later analyses.

```python
nullValue_counts=lab1_df_reduced.isnull().sum()
nullValue_counts
```

Selecting the rows that have NaNs from our reduced DataFrame:

```python
NaNsOnly=lab1_df_reduced.loc[(lab1_df_reduced['spc_latin'].isnull() ) | (lab1_df_reduced['spc_common'].isnull() ) | (lab1_df_reduced['health'].isnull() ) ]
NaNsOnly.head()
```

Accessing whether identification issues were the source of our NaNs:

```python
NaNsOnly['status'].value_counts()
```

Let's obtain the mean diameter for every tree, including those which were dead and/or unidentified.

```python
lab1_df_reduced.loc[(lab1_df_reduced['tree_dbh'] == 0) & (lab1_df_reduced['stump_diam'] == 0)]
```

Filtering out these rows and saving the new DataFrame under the name 'lab1_df_reduced_zerosFiltered':

```python
lab1_df_reduced_zerosFiltered = lab1_df_reduced.loc[(lab1_df_reduced['tree_dbh'] != 0) | (lab1_df_reduced['stump_diam'] != 0)]
```

Confirming the this filtered DataFrame is exactly 278 rows shorter than before:

```python
len(lab1_df_reduced_zerosFiltered)
```

Peform the following steps:

1) Group the rows of the filtered DataFrame by the columns 'status', 'spc_latin', 'spc_common', and 'health'.

2) Add 'dropna=False' as an argument to groupby().

3) Use mean() to get the mean tree diameter for each group.

4) Save this as a DataFrame named 'lab1_df_reduced_zerosFiltered_grouped'

Calling the name of this new DataFrame:

```python
lab1_df_reduced_zerosFiltered_grouped = lab1_df_reduced_zerosFiltered.groupby(['status','spc_latin','spc_common','health'],dropna=False).mean()
lab1_df_reduced_zerosFiltered_grouped
```

Resetting the index columns:

```python
lab1_df_reduced_zerosFiltered_grouped = lab1_df_reduced_zerosFiltered_grouped.reset_index()
lab1_df_reduced_zerosFiltered_grouped
```

Creating a new column called 'diameter' that combines the 'tree_dbh' and 'stump_diam' columns:

```python
lab1_df_reduced_zerosFiltered_grouped['diameter'] = lab1_df_reduced_zerosFiltered_grouped['tree_dbh'] + lab1_df_reduced_zerosFiltered_grouped['stump_diam']
lab1_df_reduced_zerosFiltered_grouped
```

Removing the 'tree_dbh' and 'stump_diam' columns:

```python
lab1_df_reduced_zerosFiltered_grouped.drop(columns=['tree_dbh', 'stump_diam'], inplace=True)
lab1_df_reduced_zerosFiltered_grouped
```

Assigning each row in the reduced but ungrouped DataFrame ('lab1_df_reduced') with a count of 1:

```python
lab1_df_reduced_counts=lab1_df_reduced.assign(count = 1)
lab1_df_reduced_counts.head()
```

Performing the following steps on the 'lab1_df_reduced_counts' DataFrame to produce the image below:

1) Group the rows of the filtered DataFrame by the columns 'status', 'spc_latin', 'spc_common', and 'health'.

2) Add 'dropna=False' as an argument to groupby().

3) Use sum() to add up the counts for each group.

4) Save this as a DataFrame named 'lab1_df_reduced_counts_grouped'


```python
lab1_df_reduced_counts_grouped = lab1_df_reduced_counts.groupby(['status','spc_latin','spc_common','health'],dropna=False).sum()
lab1_df_reduced_counts_grouped
```

Removing the unecessary columns and reset this DataFrame's index with this code here:

```python
lab1_df_reduced_counts_grouped.drop(columns=['tree_dbh','stump_diam'], inplace=True)
lab1_df_reduced_counts_grouped=lab1_df_reduced_counts_grouped.reset_index()
lab1_df_reduced_counts_grouped.head()
```

Producing the final dataframe named 'lab1_df_final' by left-merging* (outer='left') the two DataFrames 'lab1_df_reduced_counts_grouped' and 'lab1_df_reduced_zerosFiltered_grouped', in that order from left to right:

```python
lab1_df_final=lab1_df_reduced_counts_grouped.merge(lab1_df_reduced_zerosFiltered_grouped,how='left')
lab1_df_final
```

Making it look a bit nicer:

```python
lab1_df_final=lab1_df_final.set_index(['status','spc_latin','spc_common','health'])
lab1_df_final
```

Saving the final DataFrame as a .csv file and submitting it alongside your Google Colab notebook for grading:

```python
lab1_df_final.to_csv('/content/lab1_df_final.csv')
```
